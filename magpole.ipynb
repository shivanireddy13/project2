```python
mag = 'MAG'
ver = 'VER'
rus = 'RUS'
```

```python
quali.laps
```

```python
laps_mag = quali.laps.pick_driver(mag)
laps_ver = quali.laps.pick_driver(ver)
laps_rus = quali.laps.pick_driver(rus)
```

```python
laps_mag
```

```python
fastest_mag = laps_mag.pick_fastest()
fastest_ver = laps_ver.pick_fastest()
fastest_rus = laps_rus.pick_fastest()
fastest_mag
fastest_ver
```

```python
telemetry_mag = fastest_mag.get_telemetry().add_distance()
telemetry_ver = fastest_ver.get_telemetry().add_distance()
telemetry_rus = fastest_rus.get_telemetry().add_distance()
```

```python
team_mag = fastest_mag['Team']
team_ver = fastest_ver['Team']
team_rus = fastest_rus['Team']
```

```python
plt.rcParams['figure.figsize'] = [15,15]

fig, ax = plt.subplots(7, gridspec_kw={'height_ratios': [1, 3, 2, 1, 1, 2, 1]})

ax[0].title.set_text('plot title')

delta_time, ref_tel, compare_tel = utils.delta_time(fastest_mag, fastest_ver)

ax[0].plot(ref_tel['Distance'], delta_time)
ax[0].axhline(0)
ax[0].set(ylabel=f"Gap to {mag} (s)")

ax[1].plot(telemetry_mag['Distance'], telemetry_mag['Speed'], label=mag, color='red')
ax[1].plot(telemetry_ver['Distance'], telemetry_ver['Speed'], label=ver, color=fastf1.plotting.team_color(team_ver))
ax[1].set(ylabel='Speed')
ax[1].legend(loc="lower right")


# # Throttle trace
# ax[2].plot(telemetry_mag['Distance'], telemetry_mag['Throttle'], label=mag, color='red')
# ax[2].plot(telemetry_ver['Distance'], telemetry_ver['Throttle'], label=ver, color=fastf1.plotting.team_color(team_ver))
# ax[2].set(ylabel='Throttle')

# # Brake trace
# ax[3].plot(telemetry_mag['Distance'], telemetry_mag['Brake'], label=mag, color='red')
# ax[3].plot(telemetry_ver['Distance'], telemetry_ver['Brake'], label=ver, color=fastf1.plotting.team_color(team_ver))
# ax[3].set(ylabel='Brake')

# # Gear trace
# ax[4].plot(telemetry_mag['Distance'], telemetry_mag['nGear'], label=mag, color='red')
# ax[4].plot(telemetry_ver['Distance'], telemetry_ver['nGear'], label=ver, color=fastf1.plotting.team_color(team_ver))
# ax[4].set(ylabel='Gear')

# # RPM trace
# ax[5].plot(telemetry_mag['Distance'], telemetry_mag['RPM'], label=mag, color='red')
# ax[5].plot(telemetry_ver['Distance'], telemetry_ver['RPM'], label=ver, color=fastf1.plotting.team_color(team_ver))
# ax[5].set(ylabel='RPM')

# # DRS trace
# ax[6].plot(telemetry_mag['Distance'], telemetry_mag['DRS'], label=mag, color='red')
# ax[6].plot(telemetry_ver['Distance'], telemetry_ver['DRS'], label=ver, color=fastf1.plotting.team_color(team_ver))
# ax[6].set(ylabel='DRS')
# ax[6].set(xlabel='Lap distance (meters)')


# Hide x labels and tick labels for top plots and y ticks for right plots.
for a in ax.flat:
    a.label_outer()

#save as svg
plt.savefig('magpole.svg', format='svg', dpi=1200)
```
