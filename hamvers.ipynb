```python
%%R 

tel_ham <- read_csv('tellap7_ham.csv')
tel_ver <- read_csv('tellap7_ver.csv')
```

```python
%%R

tel_ver
```

```python
%%R
tel_combined <- rbind(tel_ver, tel_ham)
tel_combined$Driver <- c(rep("VER", nrow(tel_ver)), rep("HAM", nrow(tel_ham)))
tel_combined
```

```python
%%R

speed <- ggplot(tel_combined, aes(x = Distance, y = Speed, color = Driver)) +
  geom_line(size = 0.6) + 
  labs(title = "Lap 7") + xlab("Distance in meters") +
  scale_color_manual(name = "Driver", labels = c("VER", "HAM"), values = c("green", "red")) +
  theme_bw() +
  theme(plot.title = element_text(hjust = 0.5, size = 20)) +
  theme(plot.subtitle = element_text(hjust = 0.5, size = )) +
  theme(legend.position = "bottom") + annotate("text", x = 550, y = 300, label = "", size = 3.5, color = 'black') +
  annotate("text", x = 4150, y = 275, label = "", size = 3.5, color = "black")
```

```python
%%R -w 1000 -h 500
speed
ggsave("lap7.svg", plot = speed, width = 10, height = 5)
```
